## 3. Setup Ngrok (Port Forwarding)

**Penting:** Dapatkan ngrok auth token gratis di [ngrok.com](https://ngrok.com) lalu paste di bawah.

```python
# Paste your ngrok token here (get free at https://ngrok.com)
NGROK_TOKEN = "YOUR_NGROK_TOKEN"

from pyngrok import ngrok
ngrok.set_auth_token(NGROK_TOKEN)
```

## 4. Jalankan Flask + Ngrok

```python
import threading
import subprocess
import os

PORT = 5000
os.environ["PORT"] = str(PORT)

proc = subprocess.Popen(["python", "app.py"], env=os.environ, stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)

import time
time.sleep(15)  # Wait for Flask + model loading

tunnel = ngrok.connect(PORT)
public_url = tunnel.public_url if hasattr(tunnel, 'public_url') else str(tunnel)
print("\n" + "="*50)
print("\033[1;32mBUKA URL INI DI BROWSER:\033[0m")
print(public_url)
print("="*50)
```
